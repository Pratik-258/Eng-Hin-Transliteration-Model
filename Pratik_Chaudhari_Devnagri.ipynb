## **Reading data in required format**

```python
# Parse the XML file
tree = ET.parse("/content/drive/MyDrive/py_practise/DEVNAGRI/data/NEWS2018_M-EnHi_trn.xml")
root = tree.getroot()

# Iterate over all the Name elements
for name in root.findall("Name"):
  name_id = name.attrib["ID"]
  source_name = name.find("SourceName").text
  target_name = name.find("TargetName").text

l_data = []

for i in root.findall("Name"):
  l_data.append({"ID":i.attrib["ID"], "SourceName":i.find("SourceName").text, "TargetName":i.find("TargetName").text})
  #print(l_data)

l_data  
```

# **Creating a Tokenizer**

```python
from keras.preprocessing.text import Tokenizer

# Create a tokenizer for the source language
src_tokenizer = Tokenizer(char_level=True)
src_tokenizer.fit_on_texts([d['SourceName'] for d in l_data])

# Create a tokenizer for the target language
tgt_tokenizer = Tokenizer(char_level=True)
tgt_tokenizer.fit_on_texts([d['TargetName'] for d in l_data])

# Convert the source and target names to sequences of integers
encoded_src_names = src_tokenizer.texts_to_sequences([d['SourceName'] for d in l_data])
encoded_tgt_names = tgt_tokenizer.texts_to_sequences([d['TargetName'] for d in l_data])
```

# **Padding the text**

```python
# Find the maximum length of the target names
max_tgt_len = max(len(name) for name in encoded_tgt_names)

# Pad the source and target names to the same length
padded_src_names = pad_sequences(encoded_src_names)
padded_tgt_names = pad_sequences(encoded_tgt_names, maxlen=max_tgt_len, padding='post')
```

# **Creating Seq2Seq Model**

## **LSTM**

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, RepeatVector, TimeDistributed, Dense

# Create the seq2seq model
model = Sequential()

# Add an embedding layer for the source language
model.add(Embedding(input_dim=len(src_tokenizer.word_index) + 1,
                    output_dim=64,
                    input_length=padded_src_names.shape[1]))

# Add an LSTM layer as the encoder
model.add(LSTM(128))

# Add a repeat vector layer
model.add(RepeatVector(max_tgt_len))

# Add another LSTM layer as the decoder
model.add(LSTM(128, return_sequences=True))

# Add a TimeDistributed layer to wrap the Dense layer and produce multiple outputs, one for each time step
model.add(TimeDistributed(Dense(len(tgt_tokenizer.word_index) + 1, activation='softmax')))

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(padded_src_names, padded_tgt_names, epochs=80, batch_size=64, validation_split=0.2)
```

### **Saving the model**

```python
# Save the model
model.save("copy_seq2seq_hindi_transliteration_model.h5")
```

## **Accuracy of the model**

```python
# Evaluate the model on the test data
loss, accuracy = model.evaluate(padded_src_names, padded_tgt_names)

print('Accuracy: %f' % (accuracy * 100))
```
